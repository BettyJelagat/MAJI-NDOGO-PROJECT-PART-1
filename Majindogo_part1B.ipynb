```python
##First Create a copy of the well_pollution table
```

```python
%%sql
CREATE TABLE IF NOT EXISTS
 md_water_services.well_pollution_copy
AS ( 
    SELECT *
    FROM well_pollution
   );
```

```python
##Check the data of the copy table
```

```python
%%sql 
SELECT *
FROM well_pollution_copy
LIMIT 5; 
 
```

```python
##Update the copy's description data 
```

```python
%%sql
UPDATE well_pollution_copy
SET description = 'Bacteria: E. coli'
WHERE description = 'Clean Bacteria: E. coli';
```

```python
%%sql
UPDATE well_pollution_copy
SET description = 'Bacteria: Giardia Lamblia'
WHERE description = 'Clean Bacteria: Giardia Lamblia';
```

```python
# Lets now select from the well_pollution to see if the description is updated
```

```python
%%sql 
SELECT *
FROM well_pollution_copy
WHERE results = 'Clean' and biological > 0.01;
```

```python
#Now that it is well updated
#Next, we have to set the results with contaminated where there is a contamination in the description , and also where the biological >0.01
```

```python
%%sql
UPDATE well_pollution_copy
SET results = 'Contaminated: Biological'
WHERE biological > 0.01 AND results = 'Clean';
```

```python
#Now lets check the final results
```

```python
%%sql
SELECT *
FROM well_pollution_copy
WHERE results = "Clean" AND biological > 0.01
```

```python
#Now that the issue is solved, we can use the query to correct the orignal table then drop the copy
```

```python
%%sql
UPDATE well_pollution
SET description = 'Bacteria: E. coli'
WHERE description = 'Clean Bacteria: E. coli';


UPDATE well_pollution
SET description = 'Bacteria: Giardia Lamblia'
WHERE description = 'Clean Bacteria: Giardia Lamblia';


UPDATE well_pollution
SET results = 'Contaminated: Biological'
WHERE biological > 0.01 AND results = 'Clean';
```

```python
%%sql
SELECT *
FROM well_pollution
WHERE results = "Clean" AND biological > 0.01
```

```python
#Now dropping the copy
```

```python
%%sql
DROP TABLE well_pollution_copy;
```
